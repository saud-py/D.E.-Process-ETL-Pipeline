```python
#Print out the first 15 lines of the xml file
#print_lines(15, 'https://raw.githubusercontent.com/Sakaggi/dataengineeringtutorial/main/Extract/population_data.xml')

print_lines(15, 'population_data.xml')
```

```python
#Working with the XML file using the beautiful soup 
from bs4 import BeautifulSoup

#with open('https://raw.githubusercontent.com/Sakaggi/dataengineeringtutorial/main/Extract/population_data.xml')) as fp:

with open('population_data.xml') as fp:
  soup = BeautifulSoup(fp, "lxml" )
```

```python
#navigate through the XML document
i = 0
for record in soup.find_all('record'):
  #the find_all method gets all the fields of each record
  i += 1
  for record in record.find_all('field'):
    print(record['name'], ':', record.text)
  print()
  if i == 5:
    break
```

```python
import pandas as pd
data_dictionary = {'Country or Area':[], 'Year':[], 'Item':[], 'Value':[]}

for record in soup.find_all('record'):
    for record in record.find_all('field'):
        data_dictionary[record['name']].append(record.text)

df = pd.DataFrame.from_dict(data_dictionary)
df = df.pivot(index='Country or Area', columns='Year', values='Value')
df.reset_index(level=0, inplace=True)
```

Few pandas method can be used for transformation

1.   Melt(): It is used to change the DataFrame format from wide to long
2.   Drop(): It removes the specified rows or column
3.   merge(): It used to bring two datasets together into one, and aligning the rows from each based on common attributes
4.  sort_values(): This function is used to sorts the dataframe in Ascending or Descending order of passed column




```python
#merge the datasets together according to the instructions
#1st the melt is used to change the formatting of each data frame 

# merge the data sets together according to the instructions. First, use the melt method to change the formatting of each data frame so that it looks like this:
# Country Name, Country Code, Year, Rural Value
# Country Name, Country Code, Year, Electricity Value
df_rural_melt = pd.melt(df_rural, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],var_name = 'Year', value_name='Rural_Value')

df_electricity_melt = pd.melt(df_electricity, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='Electricity_Value')
# drop any columns from the data frames that aren't needed
df_rural_melt.drop(['Indicator Name', 'Indicator Code'], axis=1, inplace=True)

df_electricity_melt.drop(['Indicator Name', 'Indicator Code'], axis=1, inplace=True)​
# merge the data frames together based on their common columns
# in this case, the common columns are Country Name, Country Code, and Year
df_merge = df_rural_melt.merge(df_electricity_melt, how='outer',on=['Country Name', 'Country Code', 'Year'])
# sort the results by country and then by year
df_combined = df_merge.sort_values(['Country Name', 'Year'])
df_combined.head()
```
